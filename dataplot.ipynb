```python
import matplotlib.pyplot as plt

# 读取数据
with open('data.txt', 'r') as file:
    lines = file.readlines()

# 初始化三个空列表来存储每一列的数据
x = []
y = []
z = []

# 解析数据
for line in lines:
    values = line.strip().split(',')
    x.append(float(values[0]))
    y.append(float(values[1]))
    z.append(float(values[2]))

# 横坐标为索引编号
index = range(len(x))

# 绘制图形
plt.plot(index, x, label='X')
plt.plot(index, y, label='Y')
plt.plot(index, z, label='Z')

# 添加标题和标签
plt.title('Data from data.txt')
plt.xlabel('Index')
plt.ylabel('Values')

# 显示图例
plt.legend()

# 显示图形
plt.show()
```

```python
plt.plot(range(2000), x[3000:5000], label='X')
```

```python
index = range(2000)
# 绘制图形
plt.plot(index, x[3000:5000], label='X',linewidth=0.8)
plt.plot(index, y[3000:5000], label='Y',linewidth=0.8)
plt.plot(index, z[3000:5000], label='Z',linewidth=0.8)
plt.ylim((-3, 15))
plt.legend(loc=1)
plt.title(' ')
plt.xlabel(' ')
plt.ylabel(' ')
```
