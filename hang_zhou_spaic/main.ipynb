```python
# 画原始 的权重图像
def inner_plot_func(weight):

    from mpl_toolkits.axes_grid1 import make_axes_locatable

    fig, ax = plt.subplots(figsize=(5,5))

    im = ax.imshow(weight, cmap='hot_r', vmin=0, vmax=0.4)
    div = make_axes_locatable(ax)
    cax = div.append_axes("right", size="5%", pad=0.05)

    ax.set_xticks(())
    ax.set_yticks(())
    ax.set_aspect("auto")

    plt.colorbar(im, cax=cax)
    fig.tight_layout()
    plt.show()
def plot_weight_func():
    
    w_before_quant = torch.load(r"C:\Users\bignuts\Desktop\ZJU\hang_zhou\alcohol\hang_zhou_spaic\save_600\real_ysc_model_mic\parameters\_parameters_dict.pt")[r'autoname1<net>_connection1<con>:autoname1<net>_layer1<neg><-autoname1<net>_input<nod>:{weight}']
    
    w_before_quant = w_before_quant.detach().cpu().numpy()
    w_before_quant = w_before_quant.reshape(10, 10, 16, 16).transpose(0, 2, 1, 3).reshape(160, 160)
    inner_plot_func(weight=w_before_quant)
```
